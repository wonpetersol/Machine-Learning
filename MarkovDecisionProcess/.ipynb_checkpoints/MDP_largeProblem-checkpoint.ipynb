## 2. MDP Random Problem

### 2.1. MDP Random Problem (Default)

```python
P,R = mdptoolbox.example.rand(S=1000, A=2, is_sparse=False, mask=None)
```

#### 2.1.1. Random Value Iteration

```python
discount = [0.5, 0.6, 0.7, 0.8, 0.9]
for i in discount:
    vi = mdptoolbox.mdp.ValueIteration(transitions=P, reward=R, discount=i, epsilon=0.01, max_iter=1000, initial_value=0)
    vi.setVerbose()
    vi.run()
```

```python
epsilon = [0.01, 0.1, 1]
for i in epsilon:
    vi = mdptoolbox.mdp.ValueIteration(transitions=P, reward=R, discount=0.9, epsilon=i, max_iter=1000, initial_value=0)
    vi.setVerbose()
    vi.run()
```

```python
vi = mdptoolbox.mdp.ValueIteration(transitions=P, reward=R, discount=0.9, epsilon=0.01, max_iter=1000, initial_value=0)
vi.setVerbose()
vi.run()
```

```python
vi.V
```

```python
vi.policy
```

```python
vi.iter
```

```python
vi.time
```

#### 2.1.2. Ranndom Policy Iteration

```python
discount = [0.5, 0.6, 0.7, 0.8, 0.9]
for i in discount:
    pi = mdptoolbox.mdp.PolicyIterationModified(transitions=P, reward=R, discount=i, epsilon=0.01, max_iter=10)
    pi.setVerbose()
    pi.run()
```

```python
epsilon = [0.01, 0.1, 1]
for i in epsilon:
    pi = mdptoolbox.mdp.PolicyIterationModified(transitions=P, reward=R, discount=0.9, epsilon=i, max_iter=10)
    pi.setVerbose()
    pi.run()
```

```python
pi = mdptoolbox.mdp.PolicyIterationModified(transitions=P, reward=R, discount=0.9, epsilon=0.01, max_iter=10)
pi.setVerbose()
pi.run()
```

```python
pi.V
```

```python
pi.policy
```

```python
pi.iter
```

```python
pi.time
```

#### 2.1.3. Random Q-Learning

```python
discount = [0.5, 0.6, 0.7, 0.8, 0.9]
for i in discount:
    qi = mdptoolbox.mdp.QLearning(transitions=P, reward=R, discount=i, n_iter=10000)
    qi.run()
    print(qi.policy)
```

```python
iterations = [10000,100000,1000000]
for i in iterations:
    qi = mdptoolbox.mdp.QLearning(transitions=P, reward=R, discount=0.9, n_iter=i)
    qi.run()
    print(qi.policy)
```

```python
qi = mdptoolbox.mdp.QLearning(transitions=P, reward=R, discount=0.9, n_iter=1000000)
qi.run()
```

```python
qi.Q
```

```python
qi.V
```

```python
qi.policy
```

```python
qi.mean_discrepancy
```

```python
qi.time
```
